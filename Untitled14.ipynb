```python
from matplotlib import pyplot as plt
import numpy as np
plt.figure('sin&cos',
           facecolor='lightgray',
           figsize=(4,3))

plt.title('Sin&Cos',
          fontsize = 24)
plt.xlabel('sinx',fontsize = 5)
plt.ylabel('cosx',fontsize = 5)
plt.grid(linestyle = '-.')



x = np.linspace(-np.pi,np.pi,200)
print(x)
sinx = np.sin(x)
cosx = np.cos(x)/2
plt.plot(x,sinx,
         linestyle = '--',
         linewidth = 2,
         color ='orange',
         label = r'$y=sinx$')
plt.plot(x,cosx,
         linestyle='--',
         linewidth=2,
         color='red',
         label = r'$ y=\frac{1}{2}cosx$')
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
           [r'$-\pi$',r'$\frac{\pi}{2}$',r'$0$',
            r'$-\frac{\pi}{2}$',r'$\pi$'],
           fontsize = 14)


plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['bottom'].set_position(('data',0))
plt.gca().spines['left'].set_position(('data',0))

plt.legend(loc='best')

plt.scatter([np.pi/2, -np.pi/2],
            [1,-1],
            marker= 's',
            edgecolors='red',
            facecolor = 'black',
            zorder = 3)





plt.show()
```
