```python
import matplotlib.pyplot as plt
import numpy as np
import easygui as eg


# make data
x = np.arange(20, 61, 0.1)
y = []
for i in range(410):
    i = eg.enterbox('')
    y.append(float(i))


# plot
fig, ax = plt.subplots()

ax.plot(x, y)



plt.show()
```
